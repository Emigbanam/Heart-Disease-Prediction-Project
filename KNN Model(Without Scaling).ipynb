##### Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

## **Data Collection**

```python
# Get the directory where the script is located
BASE_DIR = os.getcwd()
file_path = os.path.join(BASE_DIR, "heart_disease_data.csv")

heart_df = pd.read_csv(file_path)
heart_df.head()
```

## **Data Preprocessing**

```python
#Checking For Missing Values
heart_df.isna().sum()
```

```python
#Confirming Data Types And Null Values
heart_df.info()
```

```python
#Statistical Measures
heart_df.describe()
```

```python
#Checking For The Target Distribution
print(heart_df['target'].value_counts())

heart_df['target'].value_counts().plot(kind='pie')
plt.show()
```

## **Data Splitting**

```python
#Feature-Target Splitting
X, y = heart_df.iloc[:,:-1], heart_df.iloc[:,-1]
```

```python
#Train-Test Splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=41)

X_test
```

## **Modelling**

-----
#### HyperParameter Tuning
Hyperparameter tuning helps you find the best combination of model settings that improve performance on unseen data. It ensures the model learns effectively without underfitting or overfitting. In our case we want to find the best number of neighbors for the KNN model.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

neighbor_list = list(range(5,75,5))
accuracy_list = []

for neighbor in neighbor_list:
    knn =  KNeighborsClassifier(n_neighbors=neighbor)
    predictions = knn.fit(X_train, y_train).predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    accuracy_list.append(accuracy)


results_df = pd.DataFrame({'neighbors':neighbor_list, 'accuracy':accuracy_list})
results_df.head()
```

```python
plt.plot(results_df['neighbors'], results_df['accuracy'])
plt.gca().set(xlabel='n_neighbors', ylabel='Accuracy', title='Accuracy for different n_neighbors')
plt.show()
```

The best value for n_neighbors is 60

```python
model = KNeighborsClassifier(n_neighbors=60)
model.fit(X_train, y_train)
y_preds = model.predict(X_test)
accuracy = accuracy_score(y_test, y_preds)

print("The accuracy of the KNN model is:", accuracy)
```
